```python
from sklearn.datasets import load_digits
```

```python
data = load_digits()
```

```python
X= data.data
```

```python
X.shape
X
```

```python
y= data.target
y
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
MH=DecisionTreeClassifier()
MH.fit(X,y)
print(MH.predict(X))
print(MH.score(X,y))
text_repr = tree.export_text(MH)
print(text_repr)
```
