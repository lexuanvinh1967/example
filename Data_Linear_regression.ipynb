# Lấy dữ liệu từ Dataset của Sklearn

```python
from sklearn.datasets import load_boston
X, y = load_boston(return_X_y=True)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
M = LinearRegression()
M.fit(X,y)
```

```python
y1= M.predict(X)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.plot(np.abs(y-y1))
```
